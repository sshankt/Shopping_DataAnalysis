# Data Visualization

```python
plt.figure(figsize = (14,4))
ax = sns.countplot(x = 'Age' , data = df)
ax.bar_label(ax.containers[0]) 
plt.show() 
```

```python
conditions = [
    (df['Age'] < 13),                      # Child
    (df['Age'] >= 13) & (df['Age'] <= 19), # Teenager
    (df['Age'] >= 20) & (df['Age'] <= 35), # Young Adult
    (df['Age'] > 35)                       # Adult
]
labels = ['Child', 'Teenager', 'Young', 'Adult']

# Create a new column with age group labels
df['Age Group'] = np.select(conditions, labels, default='Unknown')
```

```python
df.head() 
```

```python
plt.figure(figsize = (8,4))
ax = sns.countplot(x = 'Age Group' , data = df)
ax.bar_label(ax.containers[0]) 
plt.show() 
```

## According to the data Adult people shop the most.

```python
plt.figure(figsize = (8,4))
ax = sns.countplot(x = 'Gender' , data = df)
ax.bar_label(ax.containers[0]) 
plt.show() 
```

## Adult male people shop the most

```python
 plt.figure(figsize = (20,4))
ax = sns.countplot(x = 'Item Purchased' , data = df)
ax.bar_label(ax.containers[0]) 
plt.show() 
```

## Jewelry, Pants and Blouse are the most buying Item

```python
item = df.groupby('Item Purchased')['Purchase Amount (USD)'].sum().reset_index()
item
```

```python
plt.figure(figsize=(12, 5))
sns.barplot(
    data=item,
    x='Item Purchased',
    y='Purchase Amount (USD)',
    palette='viridis')
plt.title('Item purchased by Amount (USD)')
plt.ylabel('Purchase Amount (USD)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
 plt.figure(figsize = (8,4))
ax = sns.countplot(x = 'Category' , data = df)
ax.bar_label(ax.containers[0]) 
plt.show() 
```

```python
top_locations = df['Location'].value_counts().head(5).index
filtered_df = df[df['Location'].isin(top_locations)]

# Plot the filtered data
plt.figure(figsize=(6, 4))
ax = sns.countplot(x='Location', data=filtered_df, order=top_locations)
ax.bar_label(ax.containers[0])  # Add labels to the bars
plt.title('Top 5 Locations')
plt.show()
```
